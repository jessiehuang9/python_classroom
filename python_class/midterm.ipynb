```python
import pandas as pd
from pandas.core.frame import DataFrame
```

```python
#导入数据
report=pd.read_csv('report.txt',sep=' ')
#report=report.reindex(index=list(range(1,report.shape[0])))
report.head()
```

```python
#统计每名学生总成绩、计算平均并从高到低重新排名
data=report.drop('姓名',axis=1)
report['平均']=data.mean(axis=1)
report['总分']=data.sum(axis=1)
report=report.sort_values(by='平均',ascending=False)
report.head()
```

```python
#汇总每一科目的平均分和总平均分
report.loc['平均分']=report.mean(axis=0)
```

```python
#提取出来平均分行并删除
df=report.loc['平均分']
df=df.fillna('平均分')
df1=DataFrame(df)
df1=df1.T
report.drop(['平均分'],axis=0,inplace=True)
df1
```

```python
#把平均分放到第一行
report2=pd.concat([df1,report],ignore_index=False)
report2
```

```python
#添加名次
report2.insert(0,'名次',range(0,31))
```

```python
#替换60分以下的成绩为“不及格”
for i in range(2,12):
    report2.loc[report2.iloc[:,i] <=60, report2.columns[i]] = "不及格"
report2
```

```python
#将处理后的成绩另存为一个新文件（result.txt）
report2.to_csv('result.txt', sep=' ',index=False)
```
